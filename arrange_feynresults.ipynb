```python
import pandas as pd

results_dir = "experiments/benchmarks/feynman_physo"

# 读取benchmarks/feynman_benchmarks/FeynmanEquations.csv中Filename列
df = pd.read_csv('/home/wsq/codes/SOTA_SRs/FeynmanEquations.csv')
eq_names = df['Filename'].values
res_dic = {}
set_names = []
for v in eq_names:
    if v not in res_dic.keys() and type(v) == str:
        res_dic[v] = {
            "r2": [],
            "mae": [],
        }
        set_names.append(v)

set_names
```

```python
import pandas as pd
import os

# 文件列表
file_list = os.listdir("RES")
# 存储所有数据的列表
data_frames = []

# 读取每个csv文件并添加到列表中
for file_name in file_list:
    if file_name.endswith(".csv") and file_name != "entire.csv":
        file_path = os.path.join("RES", file_name)
        df = pd.read_csv(file_path)
        data_frames.append(df)

# 将所有数据拼接成一个DataFrame
entire_df = pd.concat(data_frames, ignore_index=True)
entire_df=entire_df.reset_index(drop=True)

# 将dataframe中的数据按照set_names中的顺序重新排列，遍历set_names，将对应的数据提取出来按顺序放到entire_df中, entire_df中的数据是按照set_names中的顺序排列的
for set_name in set_names:
    df = entire_df[entire_df['name'] == set_name]
    entire_df = entire_df[entire_df['name'] != set_name]
    entire_df = pd.concat([entire_df, df], ignore_index=True)



# 将拼接后的数据保存为sum.csv,如果存在则覆盖
output_path = 'sum.csv'
entire_df.to_csv(output_path, index=False)
```

```python
# 将"sum.csv"中的数据读取并存储到res_dic中
df = pd.read_csv(output_path)
for index, row in df.iterrows():
    res_dic[row['name']]['r2'].append(row['r2'])
    res_dic[row['name']]['mae'].append(row['mae'])
```

```python
# 遍历res_dic中的所有key，输出每个key中10-(r2的数量)
for key in res_dic.keys():
    if len(res_dic[key]['r2']) !=10:
        print(key, -10 + len(res_dic[key]['r2']))
```
